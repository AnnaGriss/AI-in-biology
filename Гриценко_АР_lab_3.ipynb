```python
# Импортируем библиотеки pandas
import pandas as pd

# Загружаем данные из CSV-файла, используем разделитель '\t', тк колонки в файле разделены табами
df = pd.read_csv('DB_3.csv', sep='\t')

# Выведем первые 5 строк, чтобы убедиться, что данные загружены корректно
df.head()
```

```python
df.info()

# Очистка в названиях колонок 
# Удаление последней колонки 'Unnamed: 15' 
df.columns = df.columns.str.replace('"', '').str.strip().str.replace('\t', ' ')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Проверка названия кoлонок после очистки
print("Очищенные названия колонок:")
print(df.columns.tolist())
```

```python
#Первые и последние строки для визуального контроля
print("Первые 5 строк:")
display(df.head())

print("\nПоследние 5 строк:")
display(df.tail())

# Основных статистические показатели для числовых признаков
print("\nОписательная статистика:")
df.describe()

# Уникальные категориальные признаки
print("\nУникальные значения в 'gender':", df['gender'].unique())
print("Уникальные значения в 'occupation':", df['occupation'].unique())
print("Уникальные значения в 'work_mode':", df['work_mode'].unique())
```

```python
# Удаление строк, где нет user_id 
df = df[df['user_id'].notna() & (df['user_id'] != '')].copy()

# Удаление дубликатов в user_id, оставив первую запись
df['user_id'] = df['user_id'].astype(str)
df = df.drop_duplicates(subset='user_id', keep='first').reset_index(drop=True)

# Разделение колонок на числовые и категориальные
categorical_cols = ['gender', 'occupation', 'work_mode']
numerical_cols = [col for col in df.columns if col not in categorical_cols + ['user_id']]

# Пропуски в категориальных признаках заменяем МОДОЙ
for col in categorical_cols:
    mode_value = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
    df[col] = df[col].fillna(mode_value)

# Пропуски в числовых признаках заменяем МЕДИАНОЙ 
for col in numerical_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)


# Проверка на оставшиеся пропуски
print("Остаточные пропуски:")
print(df.isnull().sum().sum())

# Основная статистика
print("Описательная статистика после обработки:")
df.describe()
```

```python
# Оставшиеся пропуски не являются ошибкой или артефактами 
print("Проверка успешно завершена")
```
